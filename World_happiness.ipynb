##### Happiness Score Distribution

```python
plt.figure(figsize=(24, 6))
sns.boxplot(y=df["Happiness Score"], x=df["Country"], showfliers=False)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.title("Happiness Score Distribution by Country")
plt.show()
```

##### Country with Highest Happiness Score

```python
Country_with_highest_happiness_score= happiest_country = df.loc[df["Happiness Score"].idxmax(), "Country"]
highest_happiness_score = df["Happiness Score"].max()
print(f"Country with the Highest Happiness Score is {happiest_country} ({highest_happiness_score})")
```

##### Correlation between GDP per Capital and Happiness Score

```python
correlation = df["Economy (GDP per Capita)"].corr(df["Happiness Score"])
print(f"Correlation between GDP per Capita and Happiness Score is {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Economy (GDP per Capita)"], y=df["Happiness Score"])
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.title("GDP per Capita vs Happiness Score")
plt.show()
```

##### What is the distribution of the factors that contribute to happiness?

```python
factors = ["Economy (GDP per Capita)", "Family", "Health (Life Expectancy)", "Freedom", "Trust (Government Corruption)", "Generosity"]
df[factors].hist(bins=20, figsize=(12, 8), edgecolor='black')
plt.suptitle("Distribution of Factors Contributing to Happiness")
plt.show()
```

##### Which region of the world has the highest average happiness score?

```python
region_avg_happiness = df.groupby("Region")["Happiness Score"].mean()
happiest_region = region_avg_happiness.idxmax()
highest_avg_score = region_avg_happiness.max()
print(f"Region with the Highest Average Happiness Score is {happiest_region} with ({highest_avg_score:.2f})")
```

```python
print(type(region_avg_happiness))
print(region_avg_happiness.head())
```

```python
region_avg_happiness = region_avg_happiness.to_frame().reset_index()
print(type(region_avg_happiness))
print(region_avg_happiness.head())
```

```python
plt.figure(figsize=(24, 12))
sns.barplot(data= region_avg_happiness, x="Happiness Score", y="Region", palette="viridis")
plt.xlabel("Average Happiness Score")
plt.ylabel("Region")
plt.title("Average Happiness Score by Region")
plt.show()
```
